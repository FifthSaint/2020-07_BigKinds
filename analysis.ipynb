## 빈출 단어 분석

```python
import collections
```

```python
def words_extract(str_list):
    words = list()
    for x in str_list:
        words.extend(x.split(','))
    return words
```

```python
def freq_gen(words_list, top=10, stopwords=stopwords):
    terms = [x for x in words_list if len(x) > 1 and x not in stopwords]# 한글자, 불용어 제거
    freq = collections.Counter(terms)
    sorted_freq = sorted(freq.items(), key=lambda item: item[1], reverse=True) #빈도에 따라 정렬
    return sorted_freq[:top]
```

```python
weeks_freq = list()
i = 0
top = 10

for df in weeks:
    words = words_extract(df['키워드'])
    freq = freq_gen(words)
    df_freq = pd.DataFrame(freq, columns = ['단어', '횟수'])
    filename = str(start_date+delta*i)+'_top'+str(top)+'.xlsx'
    df_freq.to_excel(filename)
    i += 1
```
